```python
# Create a starting tensor
x = torch.tensor([[1, 0, 3], [4, 5, 6]], dtype=torch.int32)
#You are telling Python to take the string "-" and repeat it 30 times.
print(f"Original Tensor:\n{x}\n" + "-"*30)

# --- FLOATING POINT TYPES ---
# Standard for DL (float32), high precision (float64), and half-precision (float16/bfloat16)
x_f32 = x.to(torch.float32)
x_f64 = x.to(torch.float64)
x_f16 = x.to(torch.float16)
x_bf16 = x.to(torch.bfloat16)

# --- INTEGER TYPES ---
# 8-bit to 64-bit signed integers
x_i8 = x.to(torch.int8)
x_i16 = x.to(torch.int16)
x_i32 = x.to(torch.int32)
x_i64 = x.to(torch.int64)

# Unsigned 8-bit (commonly used for images)
x_uint8 = x.to(torch.uint8)

# --- BOOLEAN & COMPLEX ---
x_bool = x.to(torch.bool)        # 0 becomes False, non-zero becomes True
x_c64 = x.to(torch.complex64)    # Real numbers become the 'real' part of the complex num

# --- PRINTING THE RESULTS ---
types_to_show = [
    ("32-bit Float", x_f32),
    ("64-bit Float", x_f64),
    ("16-bit Float", x_f16),
    ("BFloat16", x_bf16),
    ("8-bit Int", x_i8),
    ("32-bit Int", x_i32),
    ("64-bit Int", x_i64),
    ("Unsigned Int8", x_uint8),
    ("Boolean", x_bool),
    ("Complex 64", x_c64)
]

for name, tensor in types_to_show:
    print(f"{name:15} | Dtype: {str(tensor.dtype):18} | Example Value: {tensor[0,0]}")
```

Mathematical OPerations

```python
import torch

# Create a random 2x2 tensor
x = torch.rand(2, 2)
print(f"Original x:\n{x}\n")

# --- Math Operations ---

# addition
print(f"Addition (x + 10):\n{x + 10}")

# substraction
print(f"Subtraction (x - 5):\n{x - 5}")

# multiplication
print(f"Multiplication (x * 2):\n{x * 2}")

# division
print(f"Division (x / 2):\n{x / 2}")

# int division (floor division)
print(f"Int Division (x // 0.5):\n{x // 0.5}")

# --- Power and Modulo ---

# Power (Exponentiation)
# Use ** to raise x to a power (e.g., x squared)
power_res = x ** 2
print(f"Power (x^2):\n{power_res}")

# Modulo (Remainder) logic from your snippet
# (x // 0.5) calculates how many 0.5s fit in x, then % 2 checks if that count is even/odd
logic_gate = (x // 0.5) % 2
print(f"Logic Gate ((x // 0.5) % 2):\n{logic_gate}")
```

ElementWise


```python
# Create two random tensors of the same shape (2x3)
a = torch.rand(2, 3)
b = torch.rand(2, 3)

print("Tensor a:")
print(a)
print("\nTensor b:")
print(b)
print("-" * 30)

# 1. add
# Adds each element of a to the corresponding element of b
add_res = a + b
# Or: torch.add(a, b)

# 2. sub
# Subtracts b from a element-wise
sub_res = a - b
# Or: torch.sub(a, b)

# 3. multiply
# This is Hadamard Product (NOT matrix multiplication)
# It multiplies 0.0554 * 0.3466, etc.
mult_res = a * b
# Or: torch.mul(a, b)

# 4. division
# Divides a by b element-wise
div_res = a / b
# Or: torch.div(a, b)

# Display results
print(f"Addition Result:\n{add_res}")
print(f"Sub\n{sub_res}")
print(f"Devision\n{div_res}")
print(f"Multiplication Result:\n{mult_res}")
```

Cloning vs Copying


```python
import torch

print("--- SCENARIO 1: SIMPLE ASSIGNMENT (The 'Problem') ---")
original = torch.tensor([1, 2, 3])
copy_by_link = original  # b = a

print(f"Before change: Original = {original}, Copy = {copy_by_link}")

# We change the original tensor
original[0] = 99
print(f"After change:  Original = {original}, Copy = {copy_by_link}")
print("❌ PROBLEM: The copy changed too because they share the same memory!\n")


print("--- SCENARIO 2: CLONING (The 'Improvement') ---")
original = torch.tensor([1, 2, 3])
copy_by_clone = original.clone()  # b = a.clone()

print(f"Before change: Original = {original}, Clone = {copy_by_clone}")

# We change the original tensor
original[0] = 99
print(f"After change:  Original = {original}, Clone = {copy_by_clone}")
print("✅ SUCCESS: The clone stayed the same! It is a completely independent copy.\n")


# PROOF USING id()
print(f"ID of Original: {id(original)}")
print(f"ID of Link:     {id(copy_by_link)}  <-- Same as original")
print(f"ID of Clone:    {id(copy_by_clone)} <-- Different memory address")
```

CPU vs GPU


```python
import torch
import time

# Matrix Size
size = 10000

# 1. CPU TEST
print(f"Testing CPU for size {size}x{size}...")
cpu_matrix = torch.rand(size, size)
start_cpu = time.time()
result_cpu = torch.matmul(cpu_matrix, cpu_matrix)
end_cpu = time.time()
cpu_duration = end_cpu - start_cpu
print(f"CPU Time: {cpu_duration:.4f} seconds")

# 2. GPU TEST (if available)
if torch.cuda.is_available():
    print(f"\nTesting GPU for size {size}x{size}...")
    device = torch.device("cuda")

    # Send data to GPU
    gpu_matrix = torch.rand(size, size, device=device)

    # Warm up the GPU (first run can be slow due to memory allocation)
    _ = torch.matmul(gpu_matrix, gpu_matrix)
    torch.cuda.synchronize()

    start_gpu = time.time()
    result_gpu = torch.matmul(gpu_matrix, gpu_matrix)
    torch.cuda.synchronize() # Must synchronize to wait for GPU to finish work!
    end_gpu = time.time()
    gpu_duration = end_gpu - start_gpu

    print(f"GPU Time: {gpu_duration:.4f} seconds")

    # --- SPEEDUP CALCULATION ---
    speedup = cpu_duration / gpu_duration
    print(f"\n--- RESULTS ---")
    print(f"The GPU was {speedup:.2f}x faster than the CPU!")

else:
    print("\nCUDA (GPU) not available on this system.")
```

Reshaping, Squeezing, and NumPy

Flatten: Collapses all dimensions into one single 1D vector. Essential before passing data into a normal linear layer.
Permute: Changes the order of dimensions. Very common when converting images between different libraries (OpenCV vs PyTorch).
Unsqueeze: Adds a "fake" dimension of size 1. We use this to tell PyTorch: "This is a batch of 1 image."
Squeeze: The opposite of unsqueeze. It removes any dimension that has a size of 1 to simplify the shape.
NumPy Bridge (.numpy()): Converts the tensor to a NumPy array. Note: If the tensor is on GPU, you must call .cpu().numpy() because NumPy only runs on CPU.

```python
# --- RESHAPING ---
c = torch.rand(226, 226, 3) # Image-like tensor
print(f"Original Shape: {c.shape}")

# Flatten: Turns the whole matrix into one long 1D line of numbers
flattened = c.flatten()
print(f"Flattened Shape: {flattened.shape}")

# Permute: Reorders the dimensions (e.g., H, W, C -> C, H, W for PyTorch CNNs)
permuted = c.permute(2, 0, 1)
print(f"Permuted Shape: {permuted.shape}")

# Unsqueeze: Adds a new dimension of size 1 at index 0 (used for batching)
unsqueezed = c.unsqueeze(dim=0)
print(f"Unsqueezed Shape: {unsqueezed.shape}")

# Squeeze: Removes all '1s' from the shape
d = torch.rand(1, 10)
squeezed = d.squeeze()
print(f"Squeezed Shape: {squeezed.shape}\n")


# --- FROM TORCH TO NUMPY ---
# Use .numpy() to turn a PyTorch tensor back into a standard NumPy array
torch_tensor = torch.ones(5)
numpy_array = torch_tensor.numpy()
print(f"Object Type: {type(numpy_array)}")
```
